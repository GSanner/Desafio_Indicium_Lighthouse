### <a id='2.2.'></a>2.2. Dicionário de Dados

- **id:** Contém o identificador único dos veículos cadastrados na base de dados
- **num_fotos:** contém a quantidade de fotos que o anuncio do veículo contém
- **marca:** Contém a marca do veículo anunciado
- **modelo:** Contém o modelo do veículo anunciado
- **versao:** Contém as descrições da versão do veículo anunciando. Sua cilindrada, quantidade de válvulas, se é flex ou não, etc.
- **ano_de_fabricacao:** Contém o ano de fabricação do veículo anunciado
- **ano_modelo:** Contém o modelo do ano de fabricação do veículo anunciado

- **cambio:** Contém o tipo de câmbio do veículo anunciado
- **num_portas:** Contém a quantidade de portas do veículo anunciado
- **tipo:** Contém o tipo do veículo anunciado. Se ele é sedã, hatch, esportivo, etc.
- **blindado:** Contém informação se o veículo anunciado é blindado ou não
- **cor:** Contém a cor do veículo anunciado
- **tipo_vendedor:** Contém informações sobre o tipo do vendedor do veículo anunciado. Se é pessoa física (PF) ou se é pessoa jurídica (PJ)
- **cidade_vendedor:** Contém a cidade em que vendedor do veículo anunciado reside
- **estado_vendedor:** Contém o estado em que vendedor do veículo anunciado reside
- **anunciante:** Contém o tipo de anunciante do vendedor do veículo anunciado. Se ele é pessoa física, loja, concessionário, etc
- **entrega_delivery:** Contém informações se o vendedor faz ou não delivery do veículo anunciado
- **troca:** Contém informações o veículo anunciado já foi trocado anteriormente
- **elegivel_revisao:** Contém informações se o veículo anunciado precisa ou não de revisão
- **dono_aceita_troca:** Contém informações se o vendedor aceita ou não realizar uma troca com o veículo anunciado
- **veiculo_único_dono:** Contém informações o veículo anunciado é de um único dono
- **revisoes_concessionaria:** Contém informações se o veículo anunciado teve suas revisões feitas em concessionárias
- **ipva_pago:** Contém informações se o veículo anunciado está com o IPVA pago ou não
- **veiculo_licenciado:** Contém informações se o veículo anunciado está com o licenciamento pago ou não
- **garantia_de_fábrica:** Contém informações o veículo anunciado possui garantia de fábrica ou não
- **revisoes_dentro_agenda:** Contém informações se as revisões feitas do veículo anunciado foram realizadas dentro da agenda prevista
- **veiculo_alienado:** Contém informações se o veículo anunciado está alienado ou não
- **preco (target):** Contém as informações do preço do veículo anunciado

## <a id='3.'></a>3. Processamento dos Dados

### <a id='3.1.'></a>3.1. Ajuste da Coluna 'id'
A coluna id possui números grandes demais que dificultam a visualização dos dados. Sendo assim, vamos alterar esses números e transforma-la em indice do nosso dataframe

```python
# Criando uma nova coluna de identificadores únicos
df_cars_train['id_novo'] = df_cars_train['id'].rank(method='dense').astype(int)

# Excluindo a antiga coluna id
df_cars_train.drop('id', axis=1, inplace=True)

# Renomeando a coluna 'id_novo' para 'id'
df_cars_train.rename(columns={'id_novo': 'id'}, inplace=True)

# Transformando a coluna 'id' em indice
df_cars_train = df_cars_train.set_index('id')

# Observando o resultado
df_cars_train.head()
```

### <a id='3.2.'></a>3.2. Imputação de Dados Faltantes

Na coluna 'garantia_de_fábrica', podemos observar que apenas os carros com garantia têm informações registradas, enquanto aqueles sem garantia estão com valores nulos. Portanto, iremos realizar a imputação de valores faltantes para os carros que não possuem garantia.

```python
# Preenchendo valores nulos com 'Sem garantia'
df_cars_train['garantia_de_fábrica'].fillna('Sem garantia', inplace=True)

# Verificando
garantia_unique = df_cars_train['garantia_de_fábrica'].unique()
print(garantia_unique)
```

Na coluna 'veiculo_licenciado' ocorre o mesmo problema, sendo assim, iremos realizar a imputação de valores faltantes para os carros que não possuem licenciamento.

```python
# Preenchendo valores nulos com 'Não Licenciado'
df_cars_train['veiculo_licenciado'].fillna('Não Licenciado', inplace=True)

# Verificando
licenciado_unique = df_cars_train['veiculo_licenciado'].unique()
print(licenciado_unique)
```

Na coluna 'ipva_pago' ocorre o mesmo problema, sendo assim, iremos realizar a imputação de valores faltantes para os carros que não possuem ipva pago.

```python
# Preenchendo valores nulos com 'IPVA não pago'
df_cars_train['ipva_pago' ].fillna('IPVA não pago', inplace=True)

# Verificando
ipva_unique = df_cars_train['ipva_pago' ].unique()
print(ipva_unique)
```

Na coluna 'veiculo_único_dono' ocorre o mesmo problema, sendo assim, iremos realizar a imputação de valores faltantes para os carros que possuem mais de um dono.

```python
# Preenchendo valores nulos com 'Mais de um dono'
df_cars_train['veiculo_único_dono'].fillna('Mais de um dono', inplace=True)

# Verificando
unido_dono_unique = df_cars_train['veiculo_único_dono'].unique()
print(unido_dono_unique)
```

### <a id='3.3.'></a>3.3. Remoção de Colunas

Algumas variáveis nós não utilizaremos em nossa análise, e também não utilizaremos futuramente em nosso modelo preditivo. Sendo assim, iremos exluir algumas colunas e manter outras.

**Colunas que serão mantidas e seus motivos:**

- **marca:** A marca do veículo é um fator importante para determinar seu valor, pois reflete a qualidade, a reputação e a demanda do fabricante.
- **modelo:** O modelo do veículo também influencia seu valor, pois indica o design, o desempenho e os recursos do carro.
- **versao:** A versão do veículo pode afetar seu valor, pois mostra as especificações técnicas e os diferenciais do carro, como a cilindrada, a quantidade de válvulas, se é flex ou não, etc.
- **ano_de_fabricacao:** O ano de fabricação do veículo é um indicador de sua depreciação, ou seja, a perda de valor ao longo do tempo. Quanto mais antigo o carro, menor seu valor.
- **hodometro:** O hodômetro do veículo mede a quilometragem rodada pelo carro, o que pode indicar seu estado de conservação e uso. Quanto maior a quilometragem, menor o valor do carro.
- **cambio:** O tipo de câmbio do veículo pode influenciar seu valor, pois reflete a preferência e o conforto do motorista. Geralmente, os carros com câmbio automático são mais valorizados do que os com câmbio manual.
- **num_portas:** A quantidade de portas do veículo pode afetar seu valor, pois está relacionada ao espaço interno e à praticidade do carro. Geralmente, os carros com quatro portas são mais valorizados do que os com duas portas.
- **tipo:** O tipo do veículo pode influenciar seu valor, pois está relacionado ao estilo e à finalidade do carro. Por exemplo, os carros esportivos podem ser mais valorizados do que os sedãs ou os hatches.
- **cor:** A cor do veículo pode influenciar seu valor, pois está relacionada à estética e à demanda do mercado. Algumas cores podem ser mais populares e mais fáceis de revender do que outras.
- **ipva_pago:** O IPVA pago do veículo é um fator que pode aumentar seu valor, pois indica que o carro está em dia com suas obrigações tributárias e não possui pendências ou multas.
- **veiculo_licenciado:** O veículo licenciado é um fator que pode aumentar seu valor, pois indica que o carro está em conformidade com as normas de trânsito e segurança e pode circular livremente nas vias públicas.
- **estado_vendedor:** Será utilizado em nossa EDA para saber qual estado é melhor vender algum carro em uma característica específica.
- **garantia_de_fábrica:** Iremos utilizar para responder uma pergunta de negócio
- **veiculo_único_dono:** Iremos utilizar para responder uma pergunta de negócio
- **preco:** Nosso target

```python
# Removendo as colunas que não iremos usar
colunas_remover = ['num_fotos', 'blindado', 'ano_modelo', 'troca', 'tipo_vendedor','cidade_vendedor',
                   'anunciante', 'entrega_delivery', 'elegivel_revisao', 'veiculo_alienado',
                  'revisoes_dentro_agenda', 'revisoes_concessionaria', 'dono_aceita_troca']

df_cars_train = df_cars_train.drop(colunas_remover, axis=1, errors='ignore')
df_cars_train.head()
```

### <a id='4.2.'></a>4.2. Preço 

```python
# Arredondando o valor máximo de preço para cima para garantir que todos os dados estejam incluídos
max_preco = int(df_cars_train['preco'].max() + 20000)

# Definindo os intervalos de preços
faixas_de_preco = list(range(0, max_preco, 20000))

# Usando o método cut do pandas para categorizar os preços nos intervalos
df_cars_train['faixa_de_preco'] = pd.cut(df_cars_train['preco'], bins=faixas_de_preco, right=False)

# Calculando a frequência em cada faixa de preço
frequencia_por_faixa = df_cars_train['faixa_de_preco'].value_counts().sort_index()

# Gráfico
plt.figure(figsize=(10, 6))
frequencia_por_faixa.plot(kind='bar', width=0.8, color='skyblue')
plt.xlabel('Faixa de Preço')
plt.ylabel('Frequência')
plt.title('Frequência de Preços por Faixa')
plt.xticks(rotation=90)
plt.xlim(0, 45)
plt.show()
```

**Análise do Gráfico:** Aqui podemos ver a distribuição das vendas de carros por faixa de preço. Esse gráfico nos permite ver qual é a faixa de preço mais procurada pelos consumidores, e também como as vendas variam de acordo com o preço. Com o gráfico, fica fácil notar que a maior concentração de vendas está entre 80 mil e 100 mil, tendo uma queda relevante no número das vendas que ultrapassam esse valor.

```python
# Dados
marcas_unicas = df_cars_train['marca'].unique()

# Configurarando a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 6))


# Lista para armazenar os preços de cada marca
precos_por_marca = []

# Cores para as caixas
cores = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# Criar um gráfico de caixa para cada marca
for i, marca in enumerate(marcas_unicas):
    preco_por_marca = df_cars_train['preco'][df_cars_train['marca'] == marca]
    precos_por_marca.append(preco_por_marca)
    
    # Plotar o boxplot
    ax.boxplot(preco_por_marca, positions=[i], widths=0.6, patch_artist=True, boxprops=dict(facecolor=cores[i % len(cores)]), medianprops=dict(color='black'), labels=[marca])

# Calculando a média de todas as vendas
media = df_cars_train['preco'].mean()

# Adicionar uma linha para mostrar a média de todas as vendas
ax.axhline(y=media, color='black', linestyle='--', label='Média')

# Configurar rótulos e título
ax.set_xticks(range(len(marcas_unicas)))
ax.set_xticklabels(marcas_unicas, rotation=90)
ax.set_title("Boxplot do preço dos veículos por marca")
ax.set_xlabel("Marca")
ax.set_ylabel("Preço")

# Mostrar o gráfico
plt.legend()
plt.tight_layout()
plt.show()
```

**Análise do Gráfico:** Esse gráfico, apesar de parecer complicado, nos permite ver como o preço varia de acordo com a marca, e também quais são as marcas com vendas com os valores mais longe da média (aqui representada por uma linha preta tracejada).

### <a id='4.4.'></a>4.4. Tipo do Carro

```python
# Calculando a proporção de veículos por tipo
proporcao_por_tipo = df_cars_train['tipo'].value_counts(normalize=True)

# Grpafico
plt.figure(figsize=(12, 6))
proporcao_por_tipo.plot(kind='bar', color='skyblue')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Proporção de Veículos Vendidos (%)')
plt.title('Proporção de Veículos Vendidos por Tipo')
plt.xticks(rotation=90)

# Porcentagens
for i, proporcao in enumerate(proporcao_por_tipo):
    plt.text(i, proporcao + 0.005, f'{proporcao:.2%}', ha='center', va='bottom', fontsize=10)

plt.show()
```

Peruas, minivans e cupês basicamente não aparecem nesse gráfico, com menos de 1% cada. Isso demonstra uma procura maior por carros sedã (seguido de hatch, utilitário esportivo, e picape), que provavelmente são considerados mais versáteis, confortáveis e econômicos. Esse gráfico nos mostra que há uma preferência dos consumidores por certos tipos de carros.

##  <a id='5.'></a>5. EDA e Perguntas de Negócio

### <a id='5.1.'></a>5.1. Pergunta 1

### "Qual o melhor estado cadastrado na base de dados para se vender um carro de marca popular e por quê?"

Antes de abordarmos esta questão, é fundamental estabelecer uma definição clara do que caracteriza **carros de marcas populares**. Para isso, iremos estabelecer critérios específicos, com foco nas variáveis de **marca** e **preço**.

Nesse sentido, iniciaremos o processo segmentando as marcas mais amplamente comercializadas e, ao mesmo tempo, estabelecendo um limite superior de preço. Somente após esse filtro inicial, identificaremos os veículos que se enquadram na categoria de "populares". Com base nessa categorização, poderemos relacioná-los aos estados onde têm melhor desempenho de venda.

```python
# Calculando a participação de mercado por marca
participacao_mercado = df_cars_train['marca'].value_counts(normalize=True) * 100

marcas_porcentagem = participacao_mercado.sort_values(ascending=False)
marcas_porcentagem.head()
```

Aqui podemos notar, como foi abservado em nossa análise descritiva, as marcas mais populares. Sendo assim, utiliaremos as marcas **VOLKSWAGEN** e **CHEVROLET**, que juntas representam mais de **25%** das vendas.

```python
# Filtrando carros das marcas selecionadas e dentro do limite de preço

# Marcas Selecionadas
marcas_selecionadas = ['VOLKSWAGEN', 'CHEVROLET']

# Valor máximo de preço
valor_maximo_preco = 150000

# Filtrando
carros_populares = df_cars_train[(df_cars_train['marca'].isin(marcas_selecionadas)) & (df_cars_train['preco'] <= valor_maximo_preco)]
carros_populares.shape
```

Aqui, estabelecemos a categoria de carros populares com critérios baseados em marcas que possuem uma maior fatia de mercado e preços considerados populares. A definição dos preços foi derivada da nossa análise descritiva. Em nosso gráfico de faixa de preços, notamos um alto número de ocorrências até o valor de **150.000**; a partir desse ponto, as ocorrências começam a declinar de forma significativa.

Agora com nossos carros populares estabelecidos e separados, vamos dar continuidade.

```python
# Agrupando os Carros Populares por Estado
carros_populares_estado = carros_populares.groupby('estado_vendedor').size().reset_index(name='quantidade_de_carros')
carros_populares_estado = carros_populares_estado.sort_values(by='quantidade_de_carros', ascending=False)
print(carros_populares_estado)
```

Aqui, uma resposta óbvia se apresenta: poderíamos afirmar que o melhor estado para vender um veículo de marca popular é **São Paulo**. No entanto, como observamos anteriormente, São Paulo lidera em vendas no geral, o que pode sugerir que seja o estado mais apropriado para a venda desses veículos, simplesmente devido ao volume de transações. No entanto, uma abordagem alternativa viável seria calcular a proporção de vendas desses veículos dentro de cada estado, **proporcionando uma segunda perspectiva para essa questão**.

```python
# Calculando novamente o total de carros vendidos em cada estado
total_carros_por_estado = df_cars_train['estado_vendedor'].value_counts().reset_index()
total_carros_por_estado.columns = ['estado_vendedor', 'total_carros_vendidos']

# Mesclando o DataFrame carros_populares_estado com o total_carros_por_estado
carros_por_estado = carros_populares_estado.merge(total_carros_por_estado, on='estado_vendedor')

# Calculando a proporção de carros populares vendidos em relação ao total de carros vendidos em cada estado
carros_por_estado['proporcao_carros_populares_vendidos'] = (carros_por_estado['quantidade_de_carros'] / carros_por_estado['total_carros_vendidos']) * 100

# Classificando o DataFrame em ordem decrescente pela coluna 'proporcao_carros_populares_vendidos'
carros_por_estado = carros_por_estado.sort_values(by='proporcao_carros_populares_vendidos', ascending=False)
carros_por_estado
```

Agora, podemos notar que proporcionalmente **Rondônia** se destaca como o melhor estado para vender carros de marcas populares. Entretanto, é importante mencionar que, mesmo com uma taxa de ocorrência de **75%**, nossa amostra de Rondônia é relativamente pequena (apenas 4), o que poderia não refletir com precisão a realidade. Portanto, uma abordagem mais sólida seria considerar o segundo estado que proporcionalmente vende mais carros populares, que é a **Paraíba**, com uma taxa de ocorrência de **58%** e uma amostra um pouco mais representativa (108).

### Considerações

A escolha do melhor estado para vender carros de marca popular pode depender do critério utilizado. Se priorizarmos o volume absoluto de vendas, **São Paulo** se destaca, mas se valorizarmos uma abordagem proporcional com uma amostra representativa, a **Paraíba** pode ser uma opção promissora.

### <a id='5.2.'></a>5.2. Pergunta 2

##  <a id='6.'></a>6. EDA e Hipóteses

### <a id='6.1.'></a>6.1. Hipótese 1
### "O hodômetro é um fato que afeta o valor de um carro usado. Ou seja, quanto mais 'rodado', menos o carro vale"

A ideia desta hipótese é que quanto mais novo e menos rodado for o veículo, maior será o seu valor de mercado, pois indica que ele está em melhor estado de conservação e tem menos desgaste.

Para testar essa hipótese, vomos alcular a correlação entre o hodômetro e o valor de mercado dos carros da nossa base de dados.

Obs: Apesar da correlação não ser um teste de hipótese, ela mede a relação linear entre duas variáveis e varia de -1 a 1. Uma correlação próxima de 1 indica uma forte relação positiva entre as variáveis, enquanto uma correlação próxima de -1 indica uma forte relação negativa. Uma correlação próxima de 0 indica que não há relação linear entre as variáveis.

```python
# Selecionando apenas as colunas relevantes
hodometro_preco = df_cars_train[['hodometro', 'preco']]

# Calculando a correlação entre o hodômetro e o preço
correlacao_hodometro_preco = hodometro_preco.corr().iloc[0, 1]

print(f'A correlação entre o hodômetro e o preço é {correlacao_hodometro_preco:.2f}.')
```

### Considerações

A correlação entre o hodômetro e o valor de mercado dos carros é -0,36. Isso indica uma correlação negativa moderada entre essas duas variáveis.

Ou seja, uma correlação negativa significa que, à medida que o hodômetro aumenta (o carro percorreu mais quilômetros), o valor de mercado tende a diminuir. Isso **está em linha com a hipótese de que carros com menos quilometragem tendem a ter um valor de mercado maior, pois podem estar em melhor estado de conservação.**

### <a id='6.2.'></a>6.2. Hipótese 2
### "O fato de um veículo ser de um único dono afeta o valor dos carros usados. "

A ideia é que os compradores podem preferir veículos que tenham tido apenas um dono, pois isso pode indicar um melhor cuidado e uma menor probabilidade de problemas mecânicos ou jurídicos, afetando o seu valor.

Obs: Como aqui nossos valores na coluna 'veiculo_único_dono' não são numéricos, e possuem apenas dois grupos, iremos utilizar o p-valor do teste t de Student para verificar se há uma diferença estatisticamente significativa nas médias de preço entre os grupos.

```python
# Importando biblioteca necessária
import scipy.stats as stats

# Separando os carros em dois grupos com base na coluna 'veiculo_único_dono'
carros_unico_dono = df_cars_train[df_cars_train['veiculo_único_dono'] == 'Único dono']
carros_mais_de_um_dono = df_cars_train[df_cars_train['veiculo_único_dono'] == 'Mais de um dono']

# Calculando métricas estatísticas
media_unico_dono = carros_unico_dono['preco'].mean()
media_mais_de_um_dono = carros_mais_de_um_dono['preco'].mean()

# Ralizando um teste t
t_stat, p_valor = stats.ttest_ind(carros_unico_dono['preco'], carros_mais_de_um_dono['preco'])

print(f"Média de preço para carros com único dono: R${media_unico_dono:.2f}")
print(f"Média de preço para carros com mais de um dono: R${media_mais_de_um_dono:.2f}")
print(f"Resultado do teste t de Student (p-valor): {p_valor:.4f}")
```

### Considerações

Observando as médias dos valores, podemos ver que em geral os carros que possuem um único dono tendem a ter um **valor maior**. Além disso, o p-valor do teste t de Student é igual a 0.0000, o que indica que **há uma diferença estatisticamente significativa** nas médias de preço entre os carros com único dono e os carros com mais de um dono.

Portanto, com base nessa análise estatística, podemos concluir que o fato de um veículo ter tido apenas um dono **afeta o valor dos carros usados**.

### <a id='6.3.'></a>6.3. Hipótese 3
### "As cor de um carro é relevante para o seu valor de mercado"

Para responder a essa hipótese, além de observar visualmente as médias dos preços em relação a cor, nós iremos utilizar o teste ANOVA.

Obs: O teste ANOVA (Análise de Variância) é utilizado quando se deseja comparar a média de três ou mais grupos para determinar se existe uma diferença significativa entre eles.Como aqui temos as cores 'Branco', 'Preto', 'Prata', 'Cinza', 'Verde', e 'Vermelho' (6 grupos), optamos por esse teste.

```python
# Gráfico
plt.figure(figsize=(12, 6))
plt.hist([df_cars_train[df_cars_train['cor'] == cor]['preco'] for cor in df_cars_train['cor'].unique()], 
         bins=20, alpha=0.7, label=df_cars_train['cor'].unique())
plt.xlabel('Preço')
plt.ylabel('Contagem')
plt.title('Distribuição de Preços por Cor do Carro')
plt.legend(title='Cores', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Realizando o teste ANOVA
cores = [df_cars_train[df_cars_train['cor'] == cor]['preco'] for cor in ['Branco', 'Preto', 'Prata', 'Cinza',
                                                                         'Verde', 'Vermelho']]
resultado_anova = stats.f_oneway(*cores)
print(f"Resultado do teste ANOVA (p-valor): {resultado_anova.pvalue:.4f}")
```

### Considerações: 

Podemos analisar pelos preços médios dos carros que **a cor pode influenciar no valor do veículo**, sendo o cinza o mais valorizado. Além disso, o resultado do teste ANOVA com um valor de p igual a 0.0000 indica que **há uma diferença estatisticamente significativa entre os grupos de cores em relação aos preços dos carros**. Em outras palavras, **a cor de um carro afeta o seu valor de mercado com alta confiança estatística**.
